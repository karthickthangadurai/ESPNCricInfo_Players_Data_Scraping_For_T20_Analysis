```python
driver = webdriver.Chrome()  

# Maximize Window
driver.maximize_window() 
driver.minimize_window()  
driver.maximize_window()  
driver.switch_to.window(driver.current_window_handle)
driver.implicitly_wait(10)
#
countries_code = ['afghanistan-40','england-1','australia-2','bangladesh-25',
                  'india-6','ireland-29','new-zealand-5','pakistan-7','south-africa-3',
                  'sri-lanka-8','west-indies-4','zimbabwe-9']

country = {}

alphas = [chr(i) for i in range(97,123)]

for CounName in countries_code:
    
    country[CounName] = []
    
    for alpha in alphas:
        
        driver.get(f'https://www.espncricinfo.com/cricketers/team/{CounName}/alpha-{alpha}')

        body = driver.find_element(By.TAG_NAME,'body')

        for i in range(50):
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.2)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        players_links = [ "https://www.espncricinfo.com"+link.get('href') 
                        for link in soup.find_all('a','ds-flex ds-cursor-pointer') ]
        
        players_age = [ i.text for i in soup.find_all('span','ds-text-tight-m ds-font-regular ds-text-typo-mid3') ]
        
        filtered_players_links = []
        filtered_players_age = []
        
        for agetext,link in zip(players_age,players_links):
        
            age = re.findall(r'\d+', agetext)
            if len(age)>=1:
                if int(age[0])<=40:
                    
                    filtered_players_age.append(agetext)
                    filtered_players_links.append(link)
            
        country[CounName].append({alpha:filtered_players_links})
        
    
```

```python
#to make as a single column for all links 
import json

AppendedCountry = {}

for country in LinkFrame.columns.to_list():

    AppendedCountry[country] = []
    
    for i in range(0,26):
    
        string = LinkFrame[country][i].replace("'",'"')
        
        dictionary = json.loads(string)
        
        AppendedCountry[country].extend([i for i in dictionary.values()][0])
        
AppendedCountry
```

```python
AllPlayers = {'PlayerName':[],'PersonDetail':[],'Links':[],'TopStats':[]}

wait_time = 0
```

```python
len(AppendedCountry['zimbabwe-9'])
```

```python
#getting overview details of each player

afghanistan = AppendedCountry['zimbabwe-9']


for link in afghanistan:
    
    if link not in AllPlayers['Links']:
        #driver = webdriver.Chrome() 
        op = webdriver.ChromeOptions() #giving headless options to not show the browser
        op.add_argument('headless')
        driver = webdriver.Chrome(options=op)
        driver.get(link)
        time.sleep(1)
        
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        table = soup.find('div','ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-4 ds-mb-8')
        print(link,table)
        #personel information collected

        personnel_detail = {}

        for i in table.findAll('div'):
            val = [j.text for j in i.findAll('p')]
            
            if len(val)>=2:
                personnel_detail.update({val[0] : val[1]})

        #teams played details 
        teams_tags = table.find_next_sibling('div')
        teams_played = []
        try:
            for i in [i.text for i in teams_tags.find_all('span')]:
                if i not in teams_played:
                    teams_played.append(i)
        except:pass   
        personnel_detail.update({'Teams':[teams_played]})

        first_key = list(personnel_detail.keys())[0]  # Often gets "name"
        first_value = personnel_detail[first_key]
        
        TableNames = driver.find_elements(By.CLASS_NAME,'ds-text-title-subtle-s')
        try:
            DataFrames = pd.read_html(link)
            DataDict = {}
            
            for title,df in zip(TableNames,DataFrames):
                DataDict.update({title.text:df.to_dict()})
        except:DataDict = {}
        
        AllPlayers['TopStats'].append(DataDict)
        
        AllPlayers['PlayerName'].append(first_value)
        AllPlayers['PersonDetail'].append(personnel_detail)
        AllPlayers['Links'].append(link)

        driver.close()
        time.sleep(1) 
        wait_time = wait_time+1
        
        if wait_time%40==0:
            time.sleep(3)
      
pd.DataFrame(personnel_detail, index=[0])
    
    
```

```python
#print(soup.prettify())
pd.DataFrame(AllPlayers).to_csv(r'D:\VS_Code\EspnCricInfo\ZimbabwePlayersOverview.csv',encoding = 'utf-8-sig',index = False)
```
